<h2 style="text-align: center;"><b>Markdown</b></h2>

<img align=center src="https://upload.wikimedia.org/wikipedia/commons/thumb/4/48/Markdown-mark.svg/1280px-Markdown-mark.svg.png" style="height:90px;" width=300/>

**Markdown** - это язык разметки текста, созданный программистами для быстрого и удобного базового форматирования текста

Зачем это нужно?
1. Для добавления разметки туда, где невозможна реальная разметка. Например, в простом текстовом файле, где невозможно выделение жирным, создание заголовков, выделение цитат и пр.

2. Для более удобного написания текстов для последующей конвертации в HTML или другие форматы.

3. Во всех ноутбуках jupyter мы используем markdown (этот текст тоже написан в markdown)

Пример действия: пишете вокруг слова/фразы/любого текста по две звездочки и получаете этот текст в жирном виде

**Этот текст будет жирным**
# А вот этот текст будет большим заголовком
*Курсив*

Больше подробностей смотри здесь: [Markdown](https://github.com/adam-p/markdown-here/wiki/Markdown-Cheatsheet#code)

<h2 style="text-align: center;"><b>Latex</b></h2>

<img align=center src="https://upload.wikimedia.org/wikipedia/commons/thumb/9/92/LaTeX_logo.svg/1280px-LaTeX_logo.svg.png" style="height:90px;" width=300/>

Что же такое **Latex**? Это набор инструментов, которые из текстовых файлов, записанных с использованием специального языка разметки формирует готовые к печати документы (как правило в формате PDF). По сути, задача Latex состоит в том же, что и задача markdown, но latex позволяет значительно больше и может стать полной заменой microsoft word/pages и подобным программам. С помощью Latex во всем мире оформляются статьи, дипломные работы, книги и пр. В нем будет проще оформлять документ, содержащий какие-либо формулы/код/таблицы, ну и в целом оформление в Latex выглядит приятнее и строже, чем в ворде и подобных программах.

<h2 style="text-align: center;"><b>Git</b></h2>

<img align=center src="https://upload.wikimedia.org/wikipedia/commons/thumb/e/e0/Git-logo.svg/2000px-Git-logo.svg.png" style="height:160px;" width=300/>

Представьте, что Вам нужно сделать проект, который в процессе его создания будет нужен кому-то, кроме Вас (например, Вы хотите показать нам свои выполненные домашние задания не в Canvas) или же у Вас командная работа, и всем необходим доступ к проекту и внесенным изменениям. Можно, конечно, использовать папку на каком-нибудь облаке и дать просто кому-нибудь к ней доступ, но есть решение лучше – это **git** (не путать с github – это по сути хостинг для проектов, использующих git). И мы настоятельно рекомендуем изучить его, так как это с большой вероятностью Вам поможет в будущем.

Для начала скачиваем: https://git-scm.com/downloads
(если что, на этом же сайте есть вся документация и подробности по использованию)

Далее, создаёте аккаунт на [Github](https://github.com/) (существуют ему подобные сервисы вроде Bitbucket или GitLab, но мы будем пользоваться гитхабом)

<img align=center src="https://i1.wp.com/linuxnewbieguide.org/wp-content/uploads/2017/03/github-logo.png" style="height:90px;" width=500/>

Здесь есть отличные уроки для начинающих: https://githowto.com/ru

Еще советуем данный видеоурок: https://www.youtube.com/watch?v=HVsySz-h9r4&t=293s

Также достаточно полезным может оказаться данный материал: https://guides.github.com/introduction/flow/

Мы не будем вдаваться в тонкости git, а покажем, что вам точно понадобится в нашем курсе. Для начала определим, что такое репозиторий: это рабочая директория с Вашим проектом. По сути, это та же папка с рабочими файлами, что хранится у вас на компьютере, но находится на сервере GitHub. Поэтому вы можете работать с проектом удалённо на любой машине, не переживая, что какие-то из ваших файлов потеряются – все файлы будут в репозитории при условии, что вы их туда отправите.  Что такое форк: форк – ваша персональная версия мастер-репозитория(копия), в нём вы можете пробовать разные решения, менять код и не бояться что-то сломать в основной версии проекта.

Как сделать так, чтобы весь материал нашего курса, все ДЗ были в репозитории Вашего аккаунта, как вносить изменения и загружать выполненную ДЗ?

1. Делаете fork нашего репозитория на GitHub (кнопочка Fork в правом верхнем углу)

2. Создаёте себе папку DLSchool

3. Запускаете Git, а именно Git Bash

4. В открывшемся окне командами cd перемещаетесь в папку, которую создали в пункте 2

5. Выполняете в ней (в окне git'а) следующий набор команд:
```bash
git init*ENTER*
git pull <ссылка на ваш форк на сайте GitHub, вида .../dlschl> master*ENTER*
```

6. Теперь у Вас на компьютере в этой папке все файлы, которые есть в Вашем форке на сайте GitHub

7. Делаете ДЗ/смотрите семинары, изменяете нужные файлы, добавляете файлы и т.д.

8. Теперь Вы хотите, чтобы Ваши изменения появились на GitHub. В папке, в которой всё это лежит, выполняете:
```bash
git add .*ENTER* 
```
(точка добавляет все файлы, которые Вы изменяли; чтобы добавить только какой-то конкретный нужно написать его имя вместо точки, например: git add myfile.py)

9. Потом сразу эту команду:
```bash
git commit -m '<ваше сообщение, комментарий к проделанной работе>'*ENTER*
```

10. Далее нужно действовать так: 

    a) Если у вас на компьютере актуальная версия файлов с Вашего репозитория на GitHub, то сразу делаете:
```bash
git push <ссылка на ваш форк на сайте GitHub, вида .../dlschl> master*ENTER*
```
    b) Если Вы что-то меняли на сайте GitHub, и у Вас на компьютере неактуальная версия файлов, нет каких-то файлов, которые есть на GitHub, то сначала перед push нужно сделать pull:
```bash
git pull <ссылка на ваш форк на сайте GitHub, вида .../dlschl> master*ENTER*
```
И теперь уже команду из пункта a)

11. Изменения теперь в Вашем репозитории, можете кидать нам на почту ссылку на выполненную домашку

12. Чтобы подгрузить в Ваш форк актуальные файлы из репозитория нашей Школы (например, мы сделали новую домашку и Вы хотите её себе скачать), нужно сделать в папке DLSchool:
```bash
git pull https://github.com/miptmlschool/dlschl master
```
Git может предложить смёрджить изменения (потому что в Ваших версиях ноутбуков домашки уже сделаны), в таком случае Вам нужно вручную разрулить, какой файл выбирать (удалить файл, в котором нет Ваших изменений), если всё автоматически не разрешилось



<h2 style="text-align: center;"><b>VirtualEnv или CondaEnv</b></h2>

Представьте, что у вас много разных библиотек, и все из них имеют свои версии (одни более старую, другие более новую). Часто многие библиотеки зависят от других, и таких зависимостей может быть мног, причём в зависимости учитывается версия библиотеки. Таким образом, две разных библиотеки могут требовать двух разных версий одной и той же библиотеки (A и B зависят от Cv0.1 и Cv0.2), от которой они зависят. Что же делать?  
На помощь приходят **виртуальные окружения (virtual environments)**. Это специальные утилиты, при запуске которых активируется изолированная от системы среда, в которой лежат те и только те библиотеки и программы, которые вы туда скачаете, и именно в тех версиях в которых нужно. То есть решение проблемы в данном случае выглядит так: создаётся venv1, внутри которой библиотеки A и Cv0.1, и venv2 внутри которых библиотеки B и Cv0.2. Таким образом, вы как будто создали две мини-копии вашей файловой системы (грубо говоря), и в каждую установили нужные версии пакетов. 

Можно установить, например, виртуальное окружение в [conda](https://conda.io/docs/user-guide/tasks/manage-environments.html), или [virtualenv](https://virtualenv.pypa.io/en/stable/)
